I'm going to assign a score, of sorts, to the original dataset with the intent of quanitifying, roughly, what a desirable job is.

### Full DF Scoring

```python
min_val = min(g_df['satisfaction'])
max_val = max(g_df['satisfaction'])
diff_val = max_val - min_val
g_df['satisfaction'] = g_df['satisfaction'] -min_val

g_df['satisfaction'] = (g_df['satisfaction'] /diff_val) * 10
```

I've normalized the satisfaction column to make it more easily interpreted. a 10 in the highest-satisfaction position in this dataset.

```python
g_df['coveted_score'] = (g_df['salary'] * g_df['satisfaction']+1) / g_df['job_openings'] 
```

```python
g_df['wide_net_score'] =( (g_df['salary'] * g_df['satisfaction']+1) * g_df['job_openings'])/100000000
```

```python
covet_rank = g_df.sort_values(ascending = False, by=['coveted_score'])
```

```python
covet_rank
```

On the flipside, scoring positions by their surface-level scores to cast a wider net would give different results.

```python
wide_rank = g_df.sort_values(ascending = False, by=['wide_net_score'])
```

```python
wide_rank
```

```python
g_df.describe()
```

Between these two comparisons, I can begin to paint a picture.

Software engineers are a very frequently demanded field, based on the low satisfaction score. As years have passed there is an increase in their 'coveted' score, though it is still WELL below the mean coveted score of 176.
