```python
rs_df = rs_df[['state','state_ab', 'city','pop', 'debt', 'male_pop', 'female_pop', 'rent_mean',
               'rent_median','family_mean', 'family_median', 
               'hc_mortgage_mean', 'hc_mortgage_median', 'hc_mean','hc_median',
               'hs_degree','hs_degree_male', 'hs_degree_female', 'male_age_mean',
               'male_age_median','female_age_mean', 'female_age_median', 'hi_mean', 'hi_median',
               'pct_own', 'married', 'married_snp', 'separated', 'divorced','rent_gt_10', 'rent_gt_15', 'rent_gt_20', 'rent_gt_25',
       'rent_gt_30', 'rent_gt_35', 'rent_gt_40', 'rent_gt_50', 'home_equity']]
```

```python
rs_df.shape
```

An aid for interpreting cols:

* hc_ is housecost
* pct_owned: percent homes owned vs rented
* home_equity is % with a home equity loan
* gt_x is for rent greater than x% of income.
* hi_ is household income
* pop is population sampled, not total population
* debt is % of homes with some type of debt


```python
rs_df = rs_df.dropna()
```

```python
rs_df.shape
```

```python
rs_grp = rs_df.groupby('state').mean()
```

```python
rs_grp.head(5)
rs_grp = rs_grp.reset_index()
```

Because of the wonky nature of some of these delivered numbers, I'm going to need to perform a little more ratio-creation

```python
rs_grp['male_percent'] = rs_grp['male_pop'] / rs_grp['pop']
```

```python
rs_grp['female_percent'] = rs_grp['female_pop'] / rs_grp['pop']
```

```python
rs_grp = rs_grp.drop(columns = ['pop', 'male_pop', 'female_pop'])
```

```python
rs_grp.head(10)
```

```python
def x(a,b):
    return a - b
```

```python
rs_grp['rent_mean_to_median'] = rs_grp.apply(lambda f: x(f['rent_mean'],f['rent_median']), axis=1)
```

```python
rs_grp['family_mean_to_median'] = rs_grp.apply(lambda f: x(f['family_mean'],f['family_median']), axis=1)
```

```python
rs_grp['hc_mortgage_mean_to_median'] =rs_grp.apply(lambda f: x(f['hc_mortgage_mean'],f['hc_mortgage_median']), axis=1)

 
```

```python
rs_grp['hc_mean_to_median'] = rs_grp.apply(lambda f: x(f['hc_mean'],f['hc_median']), axis=1)
```

```python
rs_grp_corr = rs_grp.corr()
rs_grp_corr.to_csv('realestate_grouped_corr.csv')
```

I've created a file and applied a heatmap

Some notable highs and lows:

* Debt inversely correlated to age (Higher debt, lower age), it also has a decent (.7) correlation to high household income

* On that topic, a higher % of income going to rent was correlated with lower education (hs_degree)
* A low hs degree Rate was  associated more strongly with high degrees of separation(-.68) than divorce (-.22).
* The oddest correlation I saw, was male_ercent (of population) negtively correlated to rent_gt_x figures. the less males there are, the more COL issues arise (inconclusive sample-sizes and off topic)
